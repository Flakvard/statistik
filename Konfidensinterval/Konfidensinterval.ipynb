```python
df_boxplot = pd.read_excel("Data for KI middelværdi.xlsx","Opgave 4 PE ratio")
```

```python
df_boxplot.columns
```

```python
df_boxplot = df_boxplot[['Type', 'Selskab', 'P/E ratio']].copy()
```

```python
df_bank = df_boxplot.query('Type == "Bank"')
df_med = df_boxplot.query('Type == "Med"')
df_ejend = df_boxplot.query('Type == "Ejendomme"')
df_indus = df_boxplot.query('Type == "Industri"')
```

```python
fig, axes = plot.subplots(1,4,figsize=(22,5))
ax = axes[0]
ax.boxplot(df_bank['P/E ratio'], showmeans=True)
ax.grid(color='gray', linestyle='-', linewidth=0.5)
ax.set_title('Bank P/E')
ax.set_ylim(0, 70)

ax = axes[1]
ax.boxplot(df_med['P/E ratio'], showmeans=True)
ax.grid(color='gray', linestyle='-', linewidth=0.5)
ax.set_title('Med P/E')
ax.set_ylim(0, 70)

ax = axes[2]
ax.boxplot(df_ejend['P/E ratio'], showmeans=True)
ax.grid(color='gray', linestyle='-', linewidth=0.5)
ax.set_title('Ejendom P/E')
ax.set_ylim(0, 70)

ax = axes[3]
ax.boxplot(df_indus['P/E ratio'], showmeans=True)
ax.grid(color='gray', linestyle='-', linewidth=0.5)
ax.set_title('Industri P/E')
ax.set_ylim(0, 70)

plot.show()
```
