```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
df = np.random.rand(4,3)
```

```python
df
```

```python

x = 2
mean = 2.4
std_dev = 0.5
```

```python
kumlval = 1-norm.sf(x=x,loc=mean,scale=std_dev)
```

```python
kumlval
```

```python
# Define the x values to plot
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)

# Calculate the probability density function
pdf = norm.pdf(x, mean, std_dev)

# Calculate the cumulative distribution function
cdf = norm.cdf(x, mean, std_dev)

# Find the value of x where cdf is 0.211
x_fill = x[np.abs(cdf - kumlval).argmin()]

# Create the plot
fig, ax = plt.subplots()

# Plot the PDF
ax.plot(x, pdf, color='blue', label='PDF')

# Fill the area under the CDF curve
ax.fill_between(x[x<=x_fill], 0, pdf[x<=x_fill], alpha=0.5, color='orange', label='CDF')

# Set the title and axis labels
ax.set_title('Normal Distribution')
ax.set_xlabel('x')
ax.set_ylabel('Probability Density')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
