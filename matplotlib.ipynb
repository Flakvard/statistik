```python
import numpy as np
import pandas as pa
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'o--', color='purple',lw=1.5, ms=7)
# 'o--' scatter/marker plot + dashed line 
# color='purple' line color 
# lw=1.5, linewidth
# ms=7 sc markersize
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o--', color='purple',lw=1.5, ms=7)
#figure size 10x3
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o--', color='purple',lw=1.5, ms=7)
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')

#Axis labels y and x
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o--', color='purple',lw=1.5, ms=7, label='Component 1')
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.legend(loc='upper right',fontsize=10)
#Legend
# plt.plot(....., label='Component 1') adds a label to the chart
# plt.legend(loc='upper right') position for the label on the chart default lower left 
```

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)
#add two lines to the chart
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o', color='purple',lw=1.5, ms=7, label='Component 1')
plt.plot(x2,y2,'-', color='green',lw=1.5, ms=7, label='Component 2')
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.legend(loc='upper right',fontsize=10)
#copy plt.plot(......) change x and y cordinates and label
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o', color='purple',lw=1.5, ms=7, label='Data')
plt.plot(x2,y2,'-', color='green',lw=1.5, ms=7, label='Theory')
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.ylim(top=1.5) #Expand the y top limit from 1.0 to 1.5 
plt.legend(loc='upper right',fontsize=10, ncol=2) #add two columns
#Expand the y top limit from 1.0 to 1.5 
```

```python
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

```python
textstr1 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(res_a1)),
    r'$\sigma_b=%.4f$' % (np.std(res_b1))))

textstr2 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(res_a2)),
    r'$\sigma_b=%.4f$' % (np.std(res_b2))))
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3.5))
ax = axes [0]
ax.hist(res_a1, bins=30, density=True, histtype='step', label='a', color='blue')
ax.hist(res_b1, bins=30, density=True, histtype='step', label='b', color='red')
#ax.legend(loc='upper right', fancybox=False, edgecolor='black') #fancybox = Change the box to square and not rounded
ax.text(0.05,0.81, textstr1, transform=ax.transAxes, bbox = dict(facecolor='white', edgecolor='black')) #bbox change the box for text
ax.set_title('Trail 1') #Title for the chart/figure
ax.set_ylabel('Density')

#Copypaste ax after finish editing
ax = axes [1]
ax.hist(res_a2, bins=30, density=True, histtype='step', label='a', color='blue')
ax.hist(res_b2, bins=30, density=True, histtype='step', label='b', color='red')
ax.legend(loc='upper right', fancybox=False, edgecolor='black') #fancybox = Change the box to square and not rounded
ax.text(0.05,0.81, textstr2, transform=ax.transAxes, bbox = dict(facecolor='white', edgecolor='black')) #bbox change the box for text
#ax.set_ylabel('Density')
ax.set_title('Trail 2') #Title for the chart/figure
fig.text(0.5, -0.04, '$\Delta E$ [Joules]', ha='center', fontsize=20) #ha = horizontal alignment
plt.show()
```

```python
w = 3
_ = np.linspace(-3,3,100)
X, Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2,2, figsize=(7,7))
ax = axes[0][0]
ax.streamplot(X,Y,U,V)

ax = axes[0][1]
ax.streamplot(X,Y,U,V, color=speed)

ax = axes[1][0]
lw = 5*speed / speed.max()
ax.streamplot(X,Y,U,V, linewidth=lw)
```

```python
fig, axes = plt.subplots(2,2, figsize=(7,7))
ax = axes[0][0]
ax.streamplot(X,Y,U,V)

ax = axes[0][1]
ax.streamplot(X,Y,U,V, color=speed)

ax = axes[1][0]
lw = 5*speed / speed.max()
ax.streamplot(X,Y,U,V, linewidth=lw)

ax = axes[1][1]
lw = 5*speed / speed.max()
speedpoints=np.array([[0,1],[1,0]])
ax.streamplot(X,Y,U,V, start_points=speedpoints)
```
